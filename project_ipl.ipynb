```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

matches_df = pd.read_csv('ipl_matches_2008_2022.csv')
ball_df = pd.read_csv('ipl_ball_by_ball_2008_2022.csv')
```

```python
print("\nMissing values in IPL Matches data:")
print(matches_df.isnull().sum())
```

```python
matches_df.dropna(inplace=True)
```

```python
ball_df = pd.read_csv("ipl_ball_by_ball_2008_2022.csv")
```

```python
ball_df.head()
```

```python
print("\nMissing values in IPL Ball-by-Ball data:")
print(ball_df.isnull().sum())
```

```python
ball_df.dropna(inplace=True)
```

```python
ball_df.head()
```

```python
# Visual 1: Bar plot showing the number of matches played by each team
plt.figure(figsize=(12, 6))
sns.countplot(data=matches_df, x='team1', order=matches_df['team1'].value_counts().index)
plt.xticks(rotation=45)
plt.title('Number of Matches Played by Each Team')
plt.xlabel('Team')
plt.ylabel('Number of Matches')
plt.show()
```

```python
plt.figure(figsize=(8, 8))
matches_df['toss_decision'].value_counts().plot(kind='pie', autopct='%1.1f%%', colors=['skyblue', 'lightgreen'])
plt.title('Distribution of Toss Decisions')
plt.ylabel('')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(data=matches_df, y='venue', order=matches_df['venue'].value_counts().index)
plt.title('Matches Played at Each Venue')
plt.xlabel('Number of Matches')
plt.ylabel('Venue')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(matches_df['won_by'], bins=20, kde=True, color='orange')
plt.title('Distribution of Winning Margins')
plt.xlabel('Winning Margin')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=matches_df, x='player_of_match', y='won_by', order=matches_df['player_of_match'].value_counts().index[:10])
plt.xticks(rotation=45)
plt.title('Top 10 Players of the Match - Distribution of Winning Margins')
plt.xlabel('Player')
plt.ylabel('Winning Margin')
plt.show()
```

```python
plt.figure(figsize=(8, 8))
ball_df['dismisal_kind'].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Dismissals')
plt.ylabel('')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=ball_df, x='batsman_run', y='bowler', hue='batsman_run', palette='viridis')
plt.title('Batting vs Bowling Averages')
plt.xlabel('Batting Average')
plt.ylabel('Bowler')
plt.legend(title='Total Runs')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns


# Visual 8: Extras Analysis - Violin Plot
plt.figure(figsize=(12, 6))
sns.violinplot(data=ball_df, x='batting_team', y='extras_run', hue='bowler', split=True)
plt.title('Extras Conceded by Each Team')
plt.xlabel('Team')
plt.ylabel('Extras Runs')
plt.xticks(rotation=45)
plt.legend(title='Bowler')
plt.show()

# Visual 9: Innings Analysis - Line Chart
inning_runs = ball_df.groupby(['batting_team', 'innings'])['total_run'].sum().unstack()
inning_runs.T.plot(marker='o', figsize=(12, 6))
plt.title('Total Runs Scored by Each Team in Different Innings')
plt.xlabel('Innings')
plt.ylabel('Total Runs')
plt.xticks(range(1, 4), ['1st Innings', '2nd Innings', '3rd Innings'])
plt.legend(title='Team', loc='upper left')
plt.grid(True)
plt.show()

# Visual 10: Batting Performance - Box Plot
plt.figure(figsize=(12, 6))
sns.boxplot(data=ball_df, x='batting_team', y='total_run')
plt.title('Batting Performance of Each Team')
plt.xlabel('Team')
plt.ylabel('Total Runs')
plt.xticks(rotation=45)
plt.show()
```
